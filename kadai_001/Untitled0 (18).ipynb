```python
from bs4 import BeautifulSoup
import requests
url = "https://news.yahoo.co.jp/articles/a42f08601f163282f97d7ff43604589dc70b50a0"
response = requests.get(url)
soup = BeautifulSoup(response.text, "html.parser")
title_element = soup.select_one("#uamods > div.article_body.highLightSearchTarget > div > p")
title_text = title_element.text
print(title_text)
```
