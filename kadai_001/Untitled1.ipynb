```python
from openpyxl import load_workbook
from datetime import datetime
workbook = openpyxl.Workbook()
ws = workbook.active

list_information = ["株式会社ABC", "〒101-0022 東京都千代田区神田練塀町300", "TEL:03-1234-5678 FAX:03-1234-5678", "担当者名:鈴木一郎 様"]
list_header = ["商品名", "数量", "単価", "金額"]
list_summary = ["合計", "消費税", "税込合計"]
data = [
    ["商品A", 2, 10000,20000],
    ["商品B", 1, 15000, 15000]
]
today = datetime.today()
today_format = today.strftime('%Y/%m/%d')

#
ws.cell(row = 2, column = 2, value = "請求書")

#
ws.cell(row = 4, column = 6, value = "No.")
ws.cell(row = 5, column = 6, value = "日付")
ws.cell(row = 4, column = 7, value = "0001")
ws.cell(row = 5, column = 7, value = today_format)
for i in range(0, 4):
  ws.cell(row = 4 + i, column = 2, value = list_information[i])

#
start_row = 11
start_col = 2
for j in range(0, 4):
  ws.cell(row = 10, column = 2 + j, value = list_header[j])
for r, row_data in enumerate(data, start=start_row):
    for c, value in enumerate(row_data, start=start_col):
        ws.cell(row=r, column=c, value=value)
ws["E13"] = ws["E11"].value + ws["E12"].value

#
for s in range(0, 3):
  ws.cell(row = 14 + s, column = 2, value = list_summary[s])
ws["E15"] = ws["E13"].value
ws["E16"] = ws["E15"].value * 0.1
ws["E17"] = ws["E15"].value + ws["E16"].value

workbook.save("請求書_20260621.xlsx")
```
