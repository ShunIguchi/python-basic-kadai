```python
class Human:
  def __init__(self, name, age):
    self.name = name
    self.age = age

  def check_adult(self):
    if self.age >= 20:
      print(f"{self.name}は{self.age}歳で、20歳以上なので大人です。")
    else:
      print(f"{self.name}は{self.age}歳で、20以下なので大人ではありません。")


member_list = []
examples = {"AAA": 10, "BBB": 25, "CCC": 18, "DDD": 20, "EEE": 19}

for name, age in examples.items():
  member = Human(name, age)
  member_list.append(member)

for member in member_list:
  member.check_adult()
```
