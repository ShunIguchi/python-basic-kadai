```python
class Human:
  def __init__(self, name, age):
    self.name = name
    self.age = age

  def check_adult(self):
    if self.age > 20:
      print(f"{self.name}は{self.age}歳で、20歳以上なので大人です。")
    else:
      print(f"{self.name}は{self.age}歳で、20以下なので大人ではありません。")


list = []

for i in range(1, 40, 5):
  list.append(i)
  print(i)
  member = Human("Shun", i)
  member.check_adult()
```
