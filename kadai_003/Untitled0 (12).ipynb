```python
import pandas as pd
import openpyxl
from openpyxl.styles import PatternFill


# 処理対象のファイル名のリストを格納
file_list = ['2022_年間売上表.xlsx', '2023_年間売上表.xlsx']

# 結合後のデータを格納するための空のデータフレームを作成
df = pd.DataFrame()

# ファイルを一つずつ処理する
for file_name in file_list:
    # Excelファイルを読み込む
    temp_df = pd.read_excel(file_name)

    # データフレームの列名を指定
    temp_df.columns = ['売上年', '商品', '金額（千円）']

    # concat()でデータフレームを連結
    df = pd.concat([df, temp_df], ignore_index=True)

df = df.sort_values('商品', ascending = True)
df_grouped = df.groupby(['商品', '売上年']).agg({'金額（千円）':'sum'})

writer = pd.ExcelWriter('売上集計表.xlsx')
df_grouped.to_excel(writer, sheet_name='Sheet1', index=True)
writer.close()

workbook = openpyxl.load_workbook("売上集計表.xlsx")
edit = workbook.active

gray_fill = PatternFill(
    fill_type="solid",
    fgColor="F2F2F2"
)

row_num = 1

for cell in edit[row_num]:
    cell.fill = gray_fill

workbook.save('売上集計表.xlsx')

print(df_grouped)
```
