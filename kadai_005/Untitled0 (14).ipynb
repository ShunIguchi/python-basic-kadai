```python
#import libralies
import matplotlib.pyplot as plt
import numpy as np

#define plot areas
fig, axes = plt.subplots(nrows=2, ncols=3, tight_layout = True)

#linechart
x_line = np.array([0, 1, 2, 3])
y_line = np.array([1, 2, -5, 2])
axes[0, 0].plot(x_line, y_line)

#sign
x_sign = np.linspace(0, 10, 100)
y_sign = 2 + 2 * np.sin(2 * x_sign)
axes[0, 1].plot(x_sign, y_sign)

#hist
x_hist = np.random.normal(15, 10, 1000)
axes[0, 2].hist(x_hist, bins = 30)

#scatter
np.random.seed(3)
x_scatter = 4 + np.random.normal(0, 2, 60)
y_scatter = 4 + np.random.normal(0, 2, len(x_scatter))
axes[1, 0].scatter(x_scatter, y_scatter)

plt.show()
```
